```python
#importing modules
import scipy as sp
import numpy as np
import matplotlib.pyplot as pplot
import scipy.integrate as integrate
```

```python
#constants
c = 3e8
h = 6.626e-34
kb = 1.38e-23
kw = 2897 
sigma = 5.67e-8
To = 5780 
```

```python
#Planck's function in um
#Also I learned about the lambda function from Kaitlin, it lets me have 2 unknowns
B = lambda T, wavelength: 2*h*(c**2)/(((wavelength*(10**-6))**4)*wavelength*np.expm1(h*c/(kb*(wavelength*(10**-6))*T)))
plnk = lambda wavelength: B(5780, wavelength)
```

```python
#setting up x and y arrays for plotting, got help from Nathaniel for the fill_between later
x_exp = np.linspace(-2, 8, 200)
y_exp = np.linspace(-100, 20, 200)
z = B(To, 10**x_exp)
wavelen = 10**x_exp
x_max = kw/To
y_max = np.interp(x_max, 10**x_exp, z)
```

```python
#setting up arrays for tick marks
_tic = np.arange(-2, 8.01, 2)
x_ticks = 10**x_tic
y_tic = np.arange(-100, 20.01, 20)
y_ticks = 10**y_tic
```

```python
#Calculating the percentage of total emission
flux = sigma*(To**4) #max emission for a blackbody at 5780 K
ex_UV = ((integrate.quad(plnk, 0.01, 0.1)[0])*np.pi)/flux*100 #integrate between 0.01 and 0.1 and etc for below,
far_UV = ((integrate.quad(plnk, 0.1, 0.2)[0])*np.pi)/flux*100 #multiply by pi for steradian conversion and multiply by 100 for %
UVC = ((integrate.quad(plnk, 0.2, 0.28)[0])*np.pi)/flux*100
UVB = ((integrate.quad(plnk, 0.28, 0.32)[0])*np.pi)/flux*100
UVA = ((integrate.quad(plnk, 0.32, 0.4)[0])*np.pi)/flux*100
Visible = ((integrate.quad(plnk, 0.4, 0.7)[0])*np.pi)/flux*100
near_IR = ((integrate.quad(plnk, 0.7, 4)[0])*np.pi)/flux*100
thermal_IR = ((integrate.quad(plnk, 4, 50)[0])*np.pi)/flux*100
```

```python
#Plot for number 1
pplot.figure(figsize=(10, 5.5),facecolor='white')
pplot.plot(10**x_exp, z, color='red')
pplot.yscale('log')
pplot.xscale('log')
pplot.xticks(x_ticks)
pplot.yticks(y_ticks)
pplot.text(10**3, 10**-40, '% output in a given band:')
pplot.text(10**3, 10**-45, 'Extreme UV ~ '+ str(round(ex_UV, 2))+ '%')
pplot.text(10**3, 10**-50, 'Far UV ~ ' +str(round(far_UV, 2))+ '%')
pplot.text(10**3, 10**-55, 'UV-C ~ ' +str(round(UVC, 2))+ '%')
pplot.text(10**3, 10**-60, 'UV-B ~ ' +str(round(UVB, 2))+ '%')
pplot.text(10**3, 10**-65, 'UV-A ~ ' +str(round(UVA, 2))+ '%')
pplot.text(10**3, 10**-70, 'Visible ~ '+str(round(Visible, 2))+'%')
pplot.text(10**3, 10**-75, 'Near IR ~ ' +str(round(near_IR, 2))+'%')
pplot.text(10**3, 10**-80, 'Thermal IR ~ '+str(round(thermal_IR, 2))+'%')
pplot.title('BlackBody Emission curve for T=5780 K')
pplot.ylabel('$B_{λ}(T=5780)  (Wm^{-2}μm^{-1}Sr^{-1})$')
pplot.xlabel('Wavelength, \u03BB (\u03BCm)')
pplot.fill_between(wavelen, z, where=(4e-7<(wavelen*1e-6)) & ((wavelen*1e-6)<7e-7), color='lightblue' )
pplot.vlines(x=x_max, ymin=0, ymax=y_max, colors='k', linestyles='solid')
```

```python
#plot 2
pplot.figure(figsize=(10, 5.5), facecolor='white')
pplot.plot(10**x_exp, z, color='red')
pplot.xlim(0.2, 4.01)
pplot.ylim(0, 3.0e7)
pplot.title('Blackbody Emission Curve for T=5780 K')
pplot.ylabel('$B_{λ}(T=5780)  (Wm^{-2}μm^{-1}Sr^{-1})$')
pplot.xlabel('Wavelength, \u03BB (\u03BCm)')
pplot.fill_between(wavelen, z, where=(0.4<(wavelen)) & ((wavelen)<0.7), color='lightblue' )
pplot.vlines(x=x_max, ymin=0, ymax=y_max, colors='k', linestyles='solid')
```

```python
#Plot 3
pplot.figure(figsize=(10, 5.5),facecolor='white')
y2 = np.linspace(-20, 10, 5) #setting up y array for ticks (exponents)
y2_array = 10**y2 #y ticks
pplot.plot(10**x_exp, B(6000, 10**x_exp), color='purple', label = 'T=6000 K')
pplot.plot(10**x_exp, B(3000, 10**x_exp), color = 'pink', label = 'T=3000 K')
pplot.plot(10**x_exp, B(2000, 10**x_exp), color = 'red', label = 'T=2000 K')
pplot.plot(10**x_exp, B(1000, 10**x_exp), color= 'lightgreen', label= 'T=1000 K')
pplot.plot(10**x_exp, B(500, 10**x_exp), color = 'lightblue', label = 'T=500 K')
pplot.plot(10**x_exp, B(250, 10**x_exp), color= 'blue', label = 'T=250 K')
pplot.plot(10**x_exp, B(100, 10**x_exp), color='darkgreen', label= 'T=100 K')

#calulating wein's law line
max_T = [6000, 3000, 2000, 1000, 500, 250, 100]
max_T_array = np.array(max_T)
wave_T = kw/max_T_array

wave_max = B(max_T, wave_T)
pplot.plot(wave_T, wave_max, linestyle='--', color='orange')

pplot.xlabel('Wavelength, \u03BB (\u03BCm)')
pplot.ylabel('$B_{λ}(T=5780)  (Wm^{-2}μm^{-1}Sr^{-1})$')
pplot.title('Blackbody Emission Curve for Range of T')
pplot.legend()
pplot.yscale('log')
pplot.xscale('log')
pplot.yticks(y2_array)
pplot.xticks(x_ticks)
pplot.ylim(10**-20, 10**10)
```

```python
#Rayleigh-jean's approximation
RJ = lambda T, wavelength: T*2*c*kb/(((wavelength*(10**-6))**3)*wavelength)
```

```python
#Plot 4 6000 K
pplot.figure(figsize=(10, 5.5), facecolor='white')
pplot.plot(10**x_exp, B(6000, 10**x_exp), color='red')
pplot.plot(10**x_exp, RJ(6000, 10**x_exp), linestyle='--')
pplot.ylim(10**-20, 10**15)
pplot.yscale('log')
pplot.xscale('log')
pplot.title('Blackbody Emission for T=6000 K')
pplot.xlabel('Wavelength, \u03BB (\u03BCm)')
pplot.ylabel('$B_{λ}(T=5780)  (Wm^{-2}μm^{-1}Sr^{-1})$')

#setting the wavelength percentages for plot 4
#setting limits for the ymax limit
y_max_plot4 = lambda var: np.interp(var, 10**x_exp, B(6000, 10**x_exp))

#got help from Kaitlin at this section
B90_6000 = np.interp(.90, B(6000, 10**x_exp) / RJ(6000, 10**x_exp), 10**x_exp)
pplot.vlines(x=B90_6000, ymin=0, ymax=y_max_plot4(B90_6000), color='blue', label='90% \u03BB ~'+str(round((B90_6000), 0))+'\u03BCm')
B95_6000 = np.interp(.95, B(6000, 10**x_exp) / RJ(6000, 10**x_exp), 10**x_exp)
pplot.vlines(x=B95_6000, ymin=0, ymax=y_max_plot4(B95_6000), color='lightblue', label='95% \u03BB ~'+str(round((B95_6000), 0))+'\u03BCm')
B99_6000 = np.interp(.99, B(6000, 10**x_exp) / RJ(6000, 10**x_exp), 10**x_exp)
pplot.vlines(x=B99_6000, ymin=0, ymax=y_max_plot4(B99_6000), color='gray', label='99% \u03BB ~'+str(round((B99_6000), 0))+'\u03BCm')

pplot.legend()
```

```python
#Plot 5 300 K
pplot.figure(figsize=(10, 5.5), facecolor='white')
pplot.plot(10**x_exp, B(300, 10**x_exp), color='red')
pplot.plot(10**x_exp, RJ(300, 10**x_exp), linestyle='--')
pplot.ylim(10**-20, 10**10)
pplot.yscale('log')
pplot.xscale('log')
pplot.title('Blackbody Emission for T=300 K')
pplot.xlabel('Wavelength, \u03BB (\u03BCm)')
pplot.ylabel('$B_{λ}(T=5780)  (Wm^{-2}μm^{-1}Sr^{-1})$')

#setting the wavelength percentages
y_max_plot5 = lambda variable: np.interp(variable, 10**x_exp, B(300, 10**x_exp)) #set limit for striaight line
    
B90_300 = np.interp(.90, B(300, 10**x_exp) / RJ(300, 10**x_exp), 10**x_exp) #find wavelength that's 90% and etc
pplot.vlines(x=B90_300, ymin=0, ymax=y_max_plot5(B90_300), color='blue', label='90% \u03BB ~'+str(round((B90_300), 0))+'\u03BCm')
B95_300 = np.interp(.95, B(300, 10**x_exp) / RJ(300, 10**x_exp), 10**x_exp)
pplot.vlines(x=B95_300, ymin=0, ymax=y_max_plot5(B95_300), color='lightblue', label='95% \u03BB ~'+str(round((B95_300),0))+'\u03BCm')
B99_300 = np.interp(.99, B(300, 10**x_exp) / RJ(300, 10**x_exp), 10**x_exp)
pplot.vlines(x=B99_300, ymin=0, ymax=y_max_plot5(B99_300), color='gray', label='99% \u03BB ~'+str(round((B99_300),0))+' \u03BCm')

pplot.legend()

#pplot.vlines(x=kw/6000, ymin=0, ymax=B(6000, 10**x_exp), colors='k', linestyles='solid')
```
