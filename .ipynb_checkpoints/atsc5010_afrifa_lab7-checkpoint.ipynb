<b/> Problem 1: <b/>

`Consider a star with a temperature of 5780 K (temperature of the sun)` 
* Plot the intensity of radiation as a function of wavelength (from 0.01 μm to 100 m) on a log-log scale with a yrange of 10-100 to 10+20. Note—plotting over such a large wavelength range will require you to be clever how you define your variable for wavelength. You should have equal intervals in log space.

* Compute (numerically) the percentage of the total emitted radiation that falls within the following spectrums as defined on page 58 of Petty:
    1. Extreme UV
    2. Far UV
    3. UV-C
    4. UV-B
    5. UV-A
    6. Visible
    7. Near IR
    8. Thermal IR


```python
def planck(wav, T):
    h = 6.626e-34
    c = 3.0e+8
    k = 1.381e-23
    #a = 2.0*3.14*h*c**2
    a = 2.0*h*c**2
    b = h*c/(wav*1e3*1e-9*k*T)
    intensity = a/ ( ((wav*1e3*1e-9)**5) * (np.exp(b) - 1.0) )*1e-6
    return intensity

def lambda_max(T):
    return 2897/T  ### return peak wavelength in micrometer (1e-6m)

wavelengths = np.logspace(-2, 8, 10**4, endpoint=True) 

intensity5780 = planck(wavelengths, 5780.)
plt.figure(figsize=(10,5))
plt.loglog(wavelengths, intensity5780, 'r',linestyle='solid', linewidth=2)
plt.xlabel('wavelength, $\lambda$ ($\mu m$)', fontweight='bold', fontsize=12)
plt.ylabel('$B_\lambda$ (T) $[W^{-2} \mu m^{-1} Sr^{-1}]$', fontweight='bold', fontsize=12)
plt.xlim([1e-2,1e8]); plt.xticks(ticks=[1e-2,1e0,1e2,1e4,1e6,1e8], fontsize=12)
plt.ylim([1e-100,1e20]); plt.yticks(ticks=[1e-100,1e-80,1e-60,1e-40,1e-20,1e0,1e20], fontsize=12)
xmin, xmax, ymin, ymax = plt.axis()
plt.axvline(x=lambda_max(5780), ymin=ymin, ymax=0.895, color = 'k' )
plt.fill_between(x=[0.4,0.7], y1=ymin, y2=1e8,color='lightblue',alpha=.5)
plt.text(1e3,1e-30,'% output in a given band:', fontsize=12)
plt.text(1e3,1e-38,'Extreme UV ~ 0.0%', fontsize=12)
plt.text(1e3,1e-46,'Far UV ~ 0.15%', fontsize=12)
plt.text(1e3,1e-54,'UV-C ~ 2.42%', fontsize=12)
plt.text(1e3,1e-62,'UV-B ~ 1.97%', fontsize=12)
plt.text(1e3,1e-70,'UV-A ~ 7.71%', fontsize=12)
plt.text(1e3,1e-78,'Visible ~ 36.68%', fontsize=12)
plt.text(1e3,1e-86,'Near IR ~ 50.10%', fontsize=12)
plt.text(1e3,1e-94,'Thermal IR ~ 0.97%', fontsize=12)
plt.title('Black body emission curve for T = 5780')
plt.savefig('./%s.jpg' %('afrifa_lab7_blackbody-emissions_5780'),bbox_inches='tight', dpi=100)
```

<b/> Problem 3: <b/>

`The Rayleigh-Jeans Approximation is said to be valid for longer wavelengths. You can find the formula on pg. 123. Produce two separate graphs for the intensity of emitted radiation as a function of wavelength, one for a blackbody of temperature 6000 K and one of temperature 300 K. On each graph over-plot the Rayleigh-Jeans approximation as a function of wavelength. Both plots should be on a log-log scale using the same ranges as used in problem 2.`

`Calculate and indicate on the each plot the wavelength at with the Rayleigh-Jeans approximation is valid to 90%, 95% and 99%`

```python
def rayleigh(wav, T):
    c = 3.0e+8
    k = 1.381e-23
    a = 2.0*c*k
    intensity = ((a/((wav*1e3*1e-9)**4)) *T) *1e-6
    return intensity

wavelengths = np.logspace(-2, 8, 10**4, endpoint=True) 

intensity300 = planck(wavelengths, 300.)
rayleigh_intensity300 = rayleigh(wavelengths, 300.)
plt.figure(figsize=(10,5))
plt.loglog(wavelengths, intensity300, 'r')
plt.loglog(wavelengths, rayleigh_intensity300, 'g--')
plt.xlabel('wavelength, $\lambda$ ($\mu m$)', fontweight='bold', fontsize=12)
plt.ylabel('$B_\lambda$ (T) $[W^{-2} \mu m^{-1} Sr^{-1}]$', fontweight='bold', fontsize=12)
plt.xlim([1e-2,1e8]); plt.xticks(ticks=[1e-2,1e0,1e2,1e4,1e6,1e8], fontsize=12)
plt.ylim([1e-20,1e10]); plt.yticks(ticks=[1e-20,1e-15,1e-10,1e-5,1e0,1e5,1e10], fontsize=12)
xmin, xmax, ymin, ymax = plt.axis()
plt.axvline(x=231, ymin=ymin, ymax=0.56, color = 'c', label = '90% $\lambda$ ~ 231 $\mu$m')
plt.axvline(x=471, ymin=ymin, ymax=0.52, color = 'b', label = '95% $\lambda$ ~ 471 $\mu$m')
plt.axvline(x=2389, ymin=ymin, ymax=0.43, color = 'grey', label = '99% $\lambda$ ~ 2389 $\mu$m')
plt.legend()
plt.title('Black body emission curve for T = 300')
plt.savefig('./%s.jpg' %('afrifa_lab7_blackbody-emissions_300_rayleigh'),bbox_inches='tight', dpi=100)
```

```python
wavelengths = np.logspace(-2, 8, 10**4, endpoint=True) 

intensity6000 = planck(wavelengths, 6000.)
rayleigh_intensity6000 = rayleigh(wavelengths, 6000.)
plt.figure(figsize=(10,5))
plt.loglog(wavelengths, intensity6000, 'r')
plt.loglog(wavelengths, rayleigh_intensity6000, 'g--')
plt.xlabel('wavelength, $\lambda$ ($\mu m$)', fontweight='bold', fontsize=12)
plt.ylabel('$B_\lambda$ (T) $[W^{-2} \mu m^{-1} Sr^{-1}]$', fontweight='bold', fontsize=12)
plt.xlim([1e-2,1e8]); plt.xticks(ticks=[1e-2,1e0,1e2,1e4,1e6,1e8], fontsize=12)
plt.ylim([1e-20,1e10]); plt.yticks(ticks=[1e-20,1e-15,1e-10,1e-5,1e0,1e5,1e10], fontsize=12)
xmin, xmax, ymin, ymax = plt.axis()
plt.axvline(x=25, ymin=ymin, ymax=0.732, color = 'c', label = '90% $\lambda$ ~ 25 $\mu$m')
plt.axvline(x=80, ymin=ymin, ymax=0.665, color = 'b', label = '95% $\lambda$ ~ 80 $\mu$m')
plt.axvline(x=3100, ymin=ymin, ymax=0.45, color = 'grey', label = '99% $\lambda$ ~ 3100 $\mu$m')
plt.legend()
plt.title('Black body emission curve for T = 6000')
plt.savefig('./%s.jpg' %('afrifa_lab7_blackbody-emissions_6000_rayleigh'),bbox_inches='tight', dpi=100)
```
